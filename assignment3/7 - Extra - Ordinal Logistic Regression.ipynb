```r
data.raw = read.table("crabs.dat.txt", header = TRUE)
data.raw$color = factor(data.raw$color)
data.raw$y = factor(data.raw$y)
data.raw$spine = factor(data.raw$spine)
```

```r
library(MASS)

model.color <- polr(spine ~ color, data = data.raw)
summary(model.color)
```

```r
as.numeric(predict(model.color, data.raw[, 2:6]))
```

```r
require(caret)
require(boot)
options(warn=-1)


set.seed(12345)
folds = cut(seq(1,nrow(data.raw)),breaks=10,labels=FALSE)
ls.cost = 0

for (i in 1:10){
    testIndexes = which(folds==i,arr.ind=TRUE)
    train.data = data.raw[-testIndexes, ]
    test.data = data.raw[testIndexes, ]

    model = polr(spine ~ color, data = train.data[, 2:7])
    cost = mean((as.numeric(test.data[, 7]) - as.numeric(predict(model, test.data[, 2:6])))^2)
    ls.cost = ls.cost + cost

}
ls.cost/10
```
